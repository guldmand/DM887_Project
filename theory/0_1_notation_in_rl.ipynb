# Notation i Reinforcement Learning / Notation in Reinforcement Learning

En opslagstabel over de vigtigste matematiske symboler og notationer brugt i reinforcement learning.

A reference table of the most important mathematical symbols and notations used in reinforcement learning.

## Politik og værdifunktioner / Policy and Value Functions

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| $\pi$ | Politik | Policy | En afbildning fra tilstande til handlinger (eller sandsynligheder over handlinger) | A mapping from states to actions (or probabilities over actions) | $\pi(a \mid s) = P(A_t = a \mid S_t = s)$ |
| $\pi^*$ | Optimal politik | Optimal policy | Den politik der maksimerer den forventede kumulative belønning | The policy that maximizes expected cumulative reward | $\pi^* = \arg\max_\pi V^\pi(s) \;\forall s$ |

| $V^*(s)$ | Optimal tilstandsværdi | Optimal state-value | Den maksimale værdi over alle politikker | The maximum value over all policies | $V^*(s) = \max_\pi V^\pi(s)$ |
| $Q^\pi(s,a)$ | Handlingsværdifunktion | Action-value function | Forventet afkast fra tilstand $s$, handling $a$, derefter følge $\pi$ | Expected return starting from $s$, taking $a$, then following $\pi$ | $Q^\pi(s,a) = \mathbb{E}_\pi\left[\sum_{k=0}^{\infty} \gamma^k R_{t+k+1} \mid S_t = s, A_t = a\right]$ |
| $Q^*(s,a)$ | Optimal handlingsværdi | Optimal action-value | Den maksimale handlingsværdi over alle politikker | The maximum action value over all policies | $Q^*(s,a) = \max_\pi Q^\pi(s,a)$ |
| $A^\pi(s,a)$ | Fordelsfunktion | Advantage function | Forskellen mellem handlingsværdi og tilstandsværdi | Difference between action-value and state-value | $A^\pi(s,a) = Q^\pi(s,a) - V^\pi(s)$ |

## Afkast og diskontering / Returns and Discounting

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| $G_t$ | Afkast | Return | Den (diskonterede) sum af fremtidige belønninger fra tidstrin $t$ | The (discounted) sum of future rewards from time step $t$ | $G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots = \sum_{k=0}^{\infty} \gamma^k R_{t+k+1}$ |
| $\gamma$ | Diskonteringsfaktor | Discount factor | En faktor $\in [0,1]$ der afgør vægtningen af fremtidige belønninger | A factor $\in [0,1]$ that determines the weighting of future rewards | $G_t = R_{t+1} + \gamma \, G_{t+1}$ |
| $R_t$ | Belønning ved $t$ | Reward at $t$ | Belønningssignalet modtaget på tidstrin $t$ | The reward signal received at time step $t$ | $R_{t+1} = R(S_t, A_t, S_{t+1})$ |

## Overgangs- og belønningsfunktioner / Transition and Reward Functions

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| $P(s' \mid s, a)$ | Overgangssandsynlighed | Transition probability | Sandsynligheden for at nå tilstand $s'$ givet tilstand $s$ og handling $a$ | Probability of reaching state $s'$ given state $s$ and action $a$ | $P(s' \mid s, a) = \Pr(S_{t+1} = s' \mid S_t = s, A_t = a)$ |
| $R(s, a)$ | Belønningsfunktion | Reward function | Forventet belønning ved at tage handling $a$ i tilstand $s$ | Expected reward from taking action $a$ in state $s$ | $R(s,a) = \mathbb{E}[R_{t+1} \mid S_t = s, A_t = a]$ |
| $p(s', r \mid s, a)$ | Dynamikfunktion | Dynamics function | Den fulde sandsynlighedsfordeling over næste tilstand og belønning | The full probability distribution over next state and reward | $p(s', r \mid s, a) = \Pr(S_{t+1}=s', R_{t+1}=r \mid S_t=s, A_t=a)$ |

## Bellman-ligninger / Bellman Equations

Den grundlæggende Bellman-ligning siger, at værdien af en tilstand kan dekomponeres i den umiddelbare belønning plus den diskonterede værdi af næste tilstand. Alle fire varianter nedenfor er specialtilfælde af dette princip.

The fundamental Bellman equation states that the value of a state can be decomposed into the immediate reward plus the discounted value of the next state. All four variants below are special cases of this principle.

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| **Bellman-ligning** (generel) | **Bellman-ligning** | **Bellman equation** | Værdien af en tilstand er lig den umiddelbare belønning plus diskonteret værdi af efterfølgertilstanden — det grundlæggende rekursive princip i RL | The value of a state equals the immediate reward plus discounted value of the successor state — the fundamental recursive principle in RL | $V(s) = R(s) + \gamma \displaystyle\sum_{s'} P(s' \mid s)\, V(s')$ |

---

### Elementer i Bellman-ligningen / Elements of the Bellman Equation

$$V(s) = R(s) + \gamma \sum_{s'} P(s' \mid s)\, V(s')$$

| Element | Navn (DK) | Name (EN) | Betydning (DK) | Meaning (EN) |
|---------|-----------|-----------|-----------------|--------------|
| $V(s)$ | Værdifunktionen | Value function | Værdien (forventet afkast) af at befinde sig i tilstand $s$ | The value (expected return) of being in state $s$ |
| $R(s)$ | Umiddelbar belønning | Immediate reward | Den belønning agenten modtager i tilstand $s$ (her og nu) | The reward the agent receives in state $s$ (right now) |
| $\gamma$ | Diskonteringsfaktor | Discount factor | Tal mellem 0 og 1 der bestemmer hvor meget fremtidige belønninger er værd i dag. $\gamma=0$: kun nu, $\gamma=1$: fremtiden tæller lige så meget | A number between 0 and 1 determining how much future rewards are worth today. $\gamma=0$: only now, $\gamma=1$: future counts equally |
| $\sum_{s'}$ | Summation over alle næste tilstande | Sum over all next states | Vi summerer over alle mulige efterfølgertilstande $s'$ | We sum over all possible successor states $s'$ |
| $P(s' \mid s)$ | Overgangssandsynlighed | Transition probability | Sandsynligheden for at miljøet går til tilstand $s'$ givet vi er i $s$ | The probability the environment transitions to $s'$ given we are in $s$ |
| $V(s')$ | Værdien af næste tilstand | Value of next state | Det forventede afkast fra efterfølgertilstanden $s'$ (rekursion!) | The expected return from the successor state $s'$ (recursion!) |
| $\gamma \sum_{s'} P(s' \mid s)\, V(s')$ | Diskonteret forventet fremtidig værdi | Discounted expected future value | Den samlede forventede fremtidige værdi, vægtet med sandsynligheder og diskonteret | The total expected future value, weighted by probabilities and discounted |

### Konkret eksempel med tal / Worked Example with Numbers

Vi har et simpelt miljø med 3 tilstande: $s_A$, $s_B$, $s_C$. Vi vil beregne $V(s_A)$.

We have a simple environment with 3 states: $s_A$, $s_B$, $s_C$. We want to compute $V(s_A)$.

**Givne værdier / Given values:**

| Element | Værdi / Value | Forklaring (DK) | Explanation (EN) |
|---------|--------------|------------------|------------------|
| $s$ | $s_A$ | Vi beregner værdien af tilstand $s_A$ | We compute the value of state $s_A$ |
| $R(s_A)$ | $5$ | Umiddelbar belønning i $s_A$ er 5 | Immediate reward in $s_A$ is 5 |
| $\gamma$ | $0.9$ | Vi diskonterer fremtiden med faktor 0.9 | We discount the future by factor 0.9 |
| $P(s_B \mid s_A)$ | $0.6$ | 60% chance for at gå til $s_B$ | 60% chance of transitioning to $s_B$ |
| $P(s_C \mid s_A)$ | $0.4$ | 40% chance for at gå til $s_C$ | 40% chance of transitioning to $s_C$ |
| $V(s_B)$ | $10$ | Værdien af næste tilstand $s_B$ (antaget/beregnet i forvejen) | Value of next state $s_B$ (assumed/previously computed) |
| $V(s_C)$ | $2$ | Værdien af næste tilstand $s_C$ (antaget/beregnet i forvejen) | Value of next state $s_C$ (assumed/previously computed) |

**Udregning trin for trin / Step-by-step calculation:**

| Trin / Step | Beregning / Calculation | Resultat / Result | Forklaring (DK) | Explanation (EN) |
|-------------|------------------------|-------------------|------------------|------------------|
| 1. Start med ligningen | $V(s_A) = R(s_A) + \gamma \sum_{s'} P(s' \mid s_A)\, V(s')$ | — | Skriv Bellman-ligningen op | Write out the Bellman equation |
| 2. Indsæt kendte værdier | $V(s_A) = 5 + 0.9 \times \big[P(s_B \mid s_A) \cdot V(s_B) + P(s_C \mid s_A) \cdot V(s_C)\big]$ | — | Indsæt $R$, $\gamma$ og udskriv summen | Substitute $R$, $\gamma$ and expand the sum |
| 3. Indsæt sandsynligheder og værdier | $V(s_A) = 5 + 0.9 \times \big[0.6 \times 10 + 0.4 \times 2\big]$ | — | Indsæt alle tal | Plug in all numbers |
| 4. Beregn indre produkter | $V(s_A) = 5 + 0.9 \times \big[6.0 + 0.8\big]$ | — | $0.6 \times 10 = 6.0$ og $0.4 \times 2 = 0.8$ | $0.6 \times 10 = 6.0$ and $0.4 \times 2 = 0.8$ |
| 5. Beregn summen i klammer | $V(s_A) = 5 + 0.9 \times 6.8$ | $6.8$ | Summen af forventede fremtidige værdier | Sum of expected future values |
| 6. Diskontér | $V(s_A) = 5 + 6.12$ | $6.12$ | $0.9 \times 6.8 = 6.12$ = diskonteret fremtidig værdi | $0.9 \times 6.8 = 6.12$ = discounted future value |
| 7. **Facit** | $V(s_A) = 11.12$ | **$11.12$** | Umiddelbar belønning + diskonteret fremtidig værdi | Immediate reward + discounted future value |

**Opsummering / Summary:**

$$V(s_A) = \underbrace{5}_{R(s_A)} + \underbrace{0.9}_{\gamma} \times \Big[\underbrace{0.6}_{P(s_B \mid s_A)} \times \underbrace{10}_{V(s_B)} + \underbrace{0.4}_{P(s_C \mid s_A)} \times \underbrace{2}_{V(s_C)}\Big] = 5 + 6.12 = \boxed{11.12}$$

|--------|-----------|-----------|------------------|------------------|--------------------|
| Bellman-ligning for $V^\pi$ | Bellman-forventningsligning (tilstand) | Bellman expectation equation (state) | Rekursiv relation der udtrykker værdien af en tilstand via næste tilstand, under en given politik $\pi$ | Recursive relation expressing the value of a state via the next state, under a given policy $\pi$ | $V^\pi(s) = \sum_a \pi(a \mid s) \sum_{s'} P(s' \mid s,a)\left[R(s,a) + \gamma V^\pi(s')\right]$ |
| Bellman-ligning for $Q^\pi$ | Bellman-forventningsligning (handling) | Bellman expectation equation (action) | Rekursiv relation for handlingsværdifunktionen | Recursive relation for the action-value function | $Q^\pi(s,a) = R(s,a) + \gamma \sum_{s'} P(s' \mid s,a) \sum_{a'} \pi(a' \mid s') Q^\pi(s',a')$ |
| Bellman-optimalitet for $V^*$ | Bellman-optimalitetsligning (tilstand) | Bellman optimality equation (state) | Rekursiv relation for den optimale tilstandsværdi | Recursive relation for the optimal state value | $V^*(s) = \max_a \sum_{s'} P(s' \mid s,a)\left[R(s,a) + \gamma V^*(s')\right]$ |
| Bellman-optimalitet for $Q^*$ | Bellman-optimalitetsligning (handling) | Bellman optimality equation (action) | Rekursiv relation for den optimale handlingsværdi | Recursive relation for the optimal action value | $Q^*(s,a) = R(s,a) + \gamma \sum_{s'} P(s' \mid s,a) \max_{a'} Q^*(s',a')$ |

## TD-læring og opdateringsregler / TD Learning and Update Rules

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| $\alpha$ | Læringsrate | Learning rate | Skridtstørrelse for opdatering af estimater | Step size for updating estimates | $V(s) \leftarrow V(s) + \alpha \left[\delta_t\right]$ |
| $\delta_t$ | TD-fejl | TD error | Forskellen mellem TD-mål og nuværende estimat | Difference between TD target and current estimate | $\delta_t = R_{t+1} + \gamma V(S_{t+1}) - V(S_t)$ |
| $\epsilon$ | Udforskning | Exploration rate | Sandsynlighed for at vælge en tilfældig handling ($\epsilon$-grådig) | Probability of choosing a random action ($\epsilon$-greedy) | $a = \begin{cases} \arg\max_a Q(s,a) & \text{med sandsynlighed } 1 - \epsilon \\ \text{tilfældig } a & \text{med sandsynlighed } \epsilon \end{cases}$ |
| $\lambda$ | Sporstyrke | Trace decay | Aftagningsfaktor for spor i eligibility traces | Decay factor for eligibility traces | $e_t(s) = \gamma \lambda \, e_{t-1}(s) + \mathbf{1}(S_t = s)$ |
| $e_t(s)$ | Sporvektor | Eligibility trace | Holder styr på, hvor nylige og hyppige tilstande er besøgt | Tracks how recent and frequent states have been visited | $V(s) \leftarrow V(s) + \alpha \, \delta_t \, e_t(s)$ |

## Markov Decision Processes (MDPs) — endelig horisont

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel / Example |
|--------|-----------|-----------|------------------|------------------|--------------------|
| $(\mathcal{S}, \mathcal{A}, P, r, T, r_T)$ | MDP-tuplet | MDP tuple | Tilstande, handlinger, overgang, belønning, horisont, terminalbelønning | States, actions, transition, reward, horizon, terminal reward | Gridworld med $T=10$ |

| $\Pi^{HD}$ | Historie-deterministiske | History-dependent deterministic | Som HR, men deterministiske | As HR but deterministic | undermængde af HR |
| $\Pi^{MR}$ | Markov-randomiserede | Markov randomized | Politikker der kun ser nuværende tilstand $s_t$, men er randomiserede | Policies seeing only current $s_t$, randomized | $\pi_t(\cdot \mid s)$ |
| $\Pi^{MD}$ | Markov-deterministiske | Markov deterministic | Som MR, men deterministiske | As MR but deterministic | $\pi_t: \mathcal{S} \to \mathcal{A}$ |
| $\Pi^{SR}$ | Stationære randomiserede | Stationary randomized | MR med samme regel pr. tid | MR with same rule across time | bruges i $\gamma$-MDPs |
| $\Pi^{SD}$ | Stationære deterministiske | Stationary deterministic | MD med samme regel pr. tid | MD with same rule across time | bruges i $\gamma$-MDPs |

| $V_T^* = r_T$ | Boundary condition | Boundary condition | Terminal-værdi sættes lig terminalbelønning | Terminal value equals terminal reward | start på baglæns induktion |
| Bellman-optimalitet | (endelig horisont) | Finite-horizon Bellman optimality | $V_t^*(s) = \max_a \big[ r_t(s, a) + \sum_{s'} P(s' \mid s, a) V_{t+1}^*(s') \big]$ | Same | DP-rekursionen |
| Baglæns induktion | Backward induction | Backward induction | Algoritme: gå fra $t = T$ ned til $t = 0$ og fyld $V$ og $\pi$ ind tabel-vis | Algorithm: go from $t = T$ down to $t = 0$ filling $V$ and $\pi$ tablewise | $O(T \cdot \|\mathcal{S}\|^2 \cdot \|\mathcal{A}\|)$ |
| $(\mathsf{s}, \mathsf{S})$-politik | Lager-politik | $(s, S)$ inventory policy | Genopfyld op til $S$ når lager $\leq s$ | Refill to $S$ when stock $\leq s$ | klassisk i operations research |
| Tærskel-politik | Threshold policy | Threshold policy | Stop hvis $X \geq \theta_t$ | Stop if $X \geq \theta_t$ | optimal stopping |

> **Hovedresultat (forelæsning 3):** I endelig-horisont MDPs gælder
> $$\sup_{\pi \in \Pi^{HR}} V^\pi = \sup_{\pi \in \Pi^{MD}} V^\pi$$
> så vi kan altid begrænse os til **deterministiske Markov-politikker** uden tab af optimalitet — disse fås direkte fra **baglæns induktion**.

## Discounted MDPs (γ-MDPs) — operatorer og fikspunkt

**DK:** Notation indført i notebook 4 (kapitel 6). Tilføjer den **infinite-horizon** kontraktions-baserede maskinrum.

| Symbol | Navn (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel |
|--------|-----------|-----------|------------------|------------------|----------|
| $\gamma \in [0, 1)$ | Diskonteringsfaktor | Discount factor | Vægt på fremtidige belønninger; gør uendelige summer endelige | Weight on future rewards; makes infinite sums finite | $\gamma = 0.9 \Rightarrow$ effektiv horisont $\sim 10$ |

## Episodic MDPs — Stochastic Shortest Path (Forelæsning 6 / Lecture 6)

| Symbol | Name (DK) | Name (EN) | Definition (DK) | Definition (EN) | Eksempel |
|--------|-----------|-----------|-----------------|-----------------|----------|
| $\mathcal{S}_G \subseteq \mathcal{S}$ | Mål-mængde | Goal set | Ikke-tom mængde af terminale (absorberende) tilstande hvor episoden slutter | Non-empty set of terminal (absorbing) states where the episode ends | I 3×3 grid: $\mathcal{S}_G = \{8\}$ (nederste højre hjørne) |
| $\tau$ | Termineringstid | Termination / hitting time | $\tau = \inf\{t \ge 0 : s_t \in \mathcal{S}_G\}$ — tilfældig variabel | Same — random variable | $\mathbb{E}[\tau]$ er endelig under Antagelse 7.1 |
| $r_G(s)$ | Mål-belønning | Goal reward | Pr. konvention $r(s,a) = 0$ for $s \in \mathcal{S}_G$ | Same | Indlejres ofte i overgangs-belønninger fra naboer |
| Antagelse 7.1 | Properness-antagelse | Properness assumption | Alle stationære politikker er proper, eller ikke-proper politikker har $V^\pi(s) = -\infty$ | All stationary policies are proper, or non-proper ones have $V^\pi(s) = -\infty$ | Garanterer at $(I - P^\pi)^{-1}$ eksisterer |
| Proper politik | Proper policy | Proper policy | $\mathbb{E}^{\pi,s}[\tau] < \infty$ for alle $s$ | Reaches goal in finite expected time from any start | Hvis $\rho(P^\pi) < 1$, så proper |
| $V^\pi_{\text{ssp}}(s)$ | SSP værdi | SSP value | $V^\pi_{\text{ssp}}(s) = \mathbb{E}^{\pi,s}\!\big[\sum_{t=0}^{\tau-1} r(s_t, a_t)\big]$ — **ingen** $\gamma$ | Same — **no** $\gamma$ factor | Konvergens fra $\tau < \infty$ w.p. 1 |
| $V^\pi = (I - P^\pi)^{-1} r^\pi$ | Bellman matrix-eval (SSP) | Bellman matrix eval (SSP) | Eksakt løsning af $(I - P^\pi) V^\pi = r^\pi$ | Same | Substokastisk $P^\pi$ giver invertibilitet |
| $T^\pi, T^*$ | Bellman-operatorer (SSP) | Bellman operators (SSP) | $(T^\pi V)(s) = r(s,\pi) + \sum_{s'} p(s'\mid s,\pi) V(s')$; $T^*$ med $\max_a$ | Same | Ikke nødvendigvis 1-trins kontraktion |
| $(T^*)^{|\mathcal{S}|}$ | $|\mathcal{S}|$-iteration af $T^*$ | $|\mathcal{S}|$-iterate of $T^*$ | $\|(T^*)^{|\mathcal{S}|} V_1 - (T^*)^{|\mathcal{S}|} V_2\|_\infty \le (1-\epsilon)\|V_1 - V_2\|_\infty$ | Same | Sætning 7.2 i RL Foundations |
| $\epsilon$ | SSP-kontraktions-konstant | SSP contraction constant | Uniform nedre grænse for $\mathbb{P}(\tau \le |\mathcal{S}| \mid s)$, $\epsilon \in (0, 1)$ | Same | Empirisk målbar (se Eks. 6 i nb5) |
| Level-state-trick | Reduktion finite→SSP | Finite→SSP reduction | $\mathcal{S}' = \mathcal{S} \times \{0,\dots,T\}$, $\mathcal{S}_G' = \mathcal{S} \times \{T\}$ | Same | Lineær blowup $|\mathcal{S}|(T+1)$ |
| γ-trick | Reduktion discounted→SSP | Discounted→SSP reduction | Tilføj $s_G$, $p'(s_G\mid s,a) = 1-\gamma$, $p'(s'\mid s,a) = \gamma p(s'\mid s,a)$ | Same | Kun +1 ekstra tilstand; $\mathbb{E}[\tau] = 1/(1-\gamma)$ |
| $\rho(P^\pi)$ | Spektral-radius af $P^\pi$ | Spectral radius of $P^\pi$ | $\max\{\|\lambda\|: \lambda \text{ egenværdi af } P^\pi\}$ | Same | $\rho(P^\pi) < 1 \iff \pi$ proper |
| Stop-the-bus | Optimal stopping (SSP) | Optimal stopping (SSP) | $V^*(s) = \max(X_s, V^*(s+1))$ — bagudkædet | Same | Klassisk SSP-applikation (option pricing, search) |

> **Hovedresultat (forelæsning 6):** Under **Antagelse 7.1** har Bellman-ligningen $V^\pi = (I - P^\pi)^{-1} r^\pi$ en **unik løsning**, og operatoren $(T^*)^{|\mathcal{S}|}$ er en $(1-\epsilon)$-kontraktion i sup-norm. Heraf følger at både **Value Iteration** og **Policy Iteration** konvergerer til $V^*$ — **uden brug af diskonteringsfaktor**. Episodiske MDPs er det forenende rammeværk: både finite-horizon og diskonterede MDPs kan reduceres til SSP.
